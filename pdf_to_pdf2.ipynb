```python
import cv2
import pytesseract
from PIL import Image

# pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

# Load the image from file
image = cv2.imread('/Users/Sean/Downloads/chess-endgames-subset/chess-endgames-subset1024_2.jpg')

# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply binary thresholding
_, threshold_image = cv2.threshold(gray_image, 150, 255, cv2.THRESH_BINARY)

# Convert the OpenCV image to a PIL image (because pytesseract uses PIL)
pil_image = Image.fromarray(threshold_image)

# Now use pytesseract to do OCR on the image
text = pytesseract.image_to_string(pil_image, lang='eng')

print(text)

```
